#Step3
Creating a function for converting the pdf into a Zip file by using the private key which are pdfservices-api-credentials and output the zip file at the desired location

```python
def convert_pdf_to_zip(input_filename):
    print('Beginning the Extract process')

    credentials = Credentials.service_account_credentials_builder().from_file("/content/pdfservices-api-credentials.json").build()

    execution_context = ExecutionContext.create(credentials)
    extract_pdf_operation = ExtractPDFOperation.create_new()

    # Set operation input from a source file.
    source = FileRef.create_from_local_file(input_filename)
    extract_pdf_operation.set_input(source)

    # Build ExtractPDF options and set them into the operation
    extract_pdf_options: ExtractPDFOptions = ExtractPDFOptions.builder() \
        .with_elements_to_extract([ExtractElementType.TEXT, ExtractElementType.TABLES]) \
        .with_elements_to_extract_renditions([ExtractRenditionsElementType.TABLES,ExtractRenditionsElementType.FIGURES]) \
        .build()

    extract_pdf_operation.set_options(extract_pdf_options)

    # Execute the operation.
    result = extract_pdf_operation.execute(execution_context)

    # Specify the output zip file path and name
    output_zip_filename = f"/content/final_json{i}.zip"

    # Save the result to the specified location.
    result.save_as(output_zip_filename)
    print(f'Saved as {output_zip_filename}')
```

```python
import zipfile

def extract_structured_data(zip_filename, output_filename):
    with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
        for file_name in zip_ref.namelist():
            if 'structuredData' in file_name:
                zip_ref.extract(file_name, path=output_filename)
    print('Extraction completed.')
```

We are going to find all the names of the keys of the json file. for this we can put any structuresData as input as the structure of the pdfs are constant

Corresponding code in form of functions

```python
# Importing the json library
import json

# Functions for getting the keys from any json files
def extract_keys_recursive(data, keys):
    if isinstance(data, dict):
        for key in data.keys():
            keys.append(key)
            extract_keys_recursive(data[key], keys)
    elif isinstance(data, list):
        for item in data:
            extract_keys_recursive(item, keys)

def extract_keys_from_json(json_file_path):
    with open(json_file_path) as file:
        data = json.load(file)

    keys = []
    extract_keys_recursive(data, keys)
    return keys

# Example usage
json_file_path = '/content/tables0/structuredData.json'
keys = extract_keys_from_json(json_file_path)
print(set(keys))
```

```python
# Importing all the required libraries
import json
import os

# Code for removing unnecessary contents from all the json files created earlier which has inputs as path of json file
def remove_unnecessary_content(input_file):
    # Load JSON data from input file
    with open(input_file, 'r') as file:
        json_data = json.load(file)

    # Remove unnecessary content from each element
    for element in json_data['elements']:
        keys_to_keep = ['Path', 'Text','Bounds']
        element_keys = list(element.keys())
        for key in element_keys:
            if key not in keys_to_keep:
                del element[key]

    # Write the modified JSON back to the same file
    with open(input_file, 'w') as file:
        json.dump(json_data, file, indent=4)

    # Print the modified JSON data
    print(json.dumps(json_data, indent=4))

# Example usage
parent_folder = f'/content/tables{i}'  # Path to the parent folder containing the subfolders

# Iterate over subfolders in the parent folder
for folder_name in os.listdir(parent_folder):
    folder_path = os.path.join(parent_folder, folder_name)
    if os.path.isdir(folder_path):
        json_file = os.path.join(folder_path, 'structureData.json')
        if os.path.exists(json_file):
            remove_unnecessary_content(json_file)
```
